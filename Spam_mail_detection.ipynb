```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

Data Collection and Pre-Processing


```python
#loading the data from csv file to a pandas Dataframe
raw_mail_data=pd.read_csv('mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
#replace null values with a null string
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#printing the first 5 rows of the dataframe
mail_data.head()
```

```python
mail_data.shape
```

Label Ham as 1 and Spam as 0

```python
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
mail_data.head()
```

```python
#separating the data as texts and labels
X=mail_data['Message']
Y=mail_data['Category']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
print(X.shape)
print(X_test.shape)
print(X_train.shape)
```

Feature extraction

```python
#transform the text data to feature vectors that can be used as input to the logisitic regression
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=bool(True))
```

```python
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

```python
print(X_train_features)
```

```python
#convert Y_train and Y_test values as integers
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

Training the Logistic Regression Model

```python
model=LogisticRegression()
```

```python
#training the Logistic Regression Model
model.fit(X_train_features,Y_train)
```

Evaluating the Trained model

```python
#prediction on training data
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
print(accuracy_on_training_data*100)
```

```python
#prediction on test_data

prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
print(accuracy_on_test_data*100)
```

Building a Predictive System

```python
n_mail=['''Meeting tomorrow at 10 AM. Let me know if you're available.
''']
n_mail_features=feature_extraction.transform(n_mail)
prediction=model.predict(n_mail_features)
if prediction[0]==0:
  print('Spam mail')
elif prediction[0]==1:
  print('Ham mail')
```
